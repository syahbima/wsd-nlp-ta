# Create Function

```python
import pandas as pd

def list_kalimat(file_path):
  file = pd.read_csv(file_path)
  d = {}

  for index, row in file.iterrows():
    if row["kata"] not in d:
      d[row["kata"]] = [str(row["kalimat_id"])]
    else:
      d[row["kata"]].append(str(row["kalimat_id"]))
  
  return d
```

# double_annotator_agree.csv

```python
file_path = "../data/training_set/double_annotator_agree.csv"

double_annotator_agree = list_kalimat(file_path)

double_annotator_agree
```

```python
file_path = "../data/training_set/double_annotator_disagree.csv"

double_annotator_disagree = list_kalimat(file_path)

double_annotator_disagree
```
