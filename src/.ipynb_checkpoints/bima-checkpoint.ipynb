```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn import svm
```

```python
import json

with open('../data/word_to_sent.txt') as json_file:  
    word_to_sent = json.load(json_file)
    
with open('../data/single_annotator.json') as json_file:  
    id_to_sent = json.load(json_file)
    
with open('../data/list_single_annotator.json') as json_file:  
    id_to_sense = json.load(json_file)
```

```python
word_to_sent
```

```python
id_to_sense
```

```python
id_to_sent
```

```python
word_set = set()

for key in word_to_sent :
    word_set.add(key)
```

```python
word_sent = {}
label = {}

for w in word_set :
    sentences = []
    senses = []
    list_sent_id = word_to_sent[w]
    for id_sent in list_sent_id :
        try :
            sent = " ".join(id_to_sent[id_sent])
            sentences.append(sent)
            senses.append(id_to_sense[id_sent])
        except :
            pass
    word_sent[w] = sentences
    label[w] = senses        
```

```python
word_sent
```

```python
label
```

```python
bow = {}
bow_vectorizer = {}
```

```python
for w in word_sent :
    if (len((word_sent[w])) > 0):
        vectorizer = CountVectorizer()
        X = vectorizer.fit_transform(word_sent[w])
        bow[w] = X.toarray()
        bow_vectorizer[w] = vectorizer
        
```
